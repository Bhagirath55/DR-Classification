```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a retina image (replace with your actual image path)
image_path = r"C:\Users\bhagi\Desktop\AI_ML_Projects\Early_Detection_of_Diabetic_Retinopathy\dataset\4\1a7e3356b39c.jpg"  # Change this to your actual image
retina_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display original image
plt.figure(figsize=(6, 6))
plt.title("Original Retina Image")
plt.imshow(retina_img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
from PyWavelets import pywt 

# Apply Wavelet Transform
coeffs2 = pywt.dwt2(retina_img, 'haar')
LL, (LH, HL, HH) = coeffs2

# Display wavelet components
fig, axes = plt.subplots(1, 4, figsize=(12, 4))
axes[0].imshow(LL, cmap="gray")
axes[0].set_title("LL (Approximation)")

axes[1].imshow(LH, cmap="gray")
axes[1].set_title("LH (Horizontal Edges)")

axes[2].imshow(HL, cmap="gray")
axes[2].set_title("HL (Vertical Edges)")

axes[3].imshow(HH, cmap="gray")
axes[3].set_title("HH (Diagonal Edges)")

for ax in axes:
    ax.axis("off")

plt.show()
```
